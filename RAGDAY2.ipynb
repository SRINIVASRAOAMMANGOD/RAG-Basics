```python
documents = [
    "Python is a programming language",
    "Django is a web framework",
    "Kerala is a state in India",
    "Machine learning uses data"
]
```

Loads the embedding model.

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2")

doc_embeddings = model.encode(documents)
```

Explanation

doc_embeddings = model.encode(documents)

Converts all 4 documents into vectors.

Create a Query

```python
query = "What is Django?"
query_embedding = model.encode(query)
```

Explanation

query_embedding

Converts the question into a vector.

Compare Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity

scores = cosine_similarity(
    [query_embedding],
    doc_embeddings
)

print(scores)
```

Explanation

Compares:

"What is Django?"

with every document.

Find Best Match

```python
best_index = scores.argmax()

print(documents[best_index])
```

Explanation

argmax()

returns the index of the highest similarity score.

WorkFlow

Documents

↓

Embeddings

↓

Query Embedding

↓

Cosine Similarity

↓

Best Match

Create Database

```python
import chromadb

client = chromadb.Client()

collection = client.create_collection(
    name="my_docs"
)
```

Explanation

client = chromadb.Client()

Starts ChromaDB.

---

collection = client.create_collection(...)

Creates a collection (like a SQL table).

Add Documents

```python
collection.add(
    documents=[
        "Python is a programming language",
        "Django is a web framework",
        "Kerala is a state in India"
    ],
    ids=["1", "2", "3"]
)
```

Explanation

documents=[...]

Data to store.

---


ids=[...]

Unique IDs.

Chroma automatically creates embeddings.


---

Search

```python
results = collection.query(
    query_texts=["What is Django?"],
    n_results=1
)

print(results)
```

Explanation

query_texts=[...]

Your question.

---

n_results=1

Return best match.
