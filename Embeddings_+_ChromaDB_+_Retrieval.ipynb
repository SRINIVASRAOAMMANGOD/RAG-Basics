```python
# Upload PDF
from google.colab import files

uploaded = files.upload()
```

```python
# Load PDF
from langchain_community.document_loaders import PyPDFLoader

pdf_file = list(uploaded.keys())[0]

loader = PyPDFLoader(pdf_file)
documents = loader.load()

print("Pages:", len(documents))
```

```python
# Combine all pages into one text
text = ""

for doc in documents:
    text += doc.page_content + "\n"

print("Characters:", len(text))
```

```python
# Chunking
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)

chunks = text_splitter.split_text(text)

print("Number of chunks:", len(chunks))
```

```python
# Preview first 5 chunks
for i, chunk in enumerate(chunks[:5]):
    print(f"\n----- Chunk {i+1} -----\n")
    print(chunk)
```

```python
from sentence_transformers import SentenceTransformer

embedding_model = SentenceTransformer("all-MiniLM-L6-v2")

embeddings = embedding_model.encode(chunks)

print(embeddings.shape)
```

```python
!pip install -q chromadb

import chromadb

client = chromadb.Client()

collection = client.create_collection("rag_collection")

collection.add(
    embeddings=embeddings.tolist(),
    documents=chunks,
    ids=[f"chunk_{i}" for i in range(len(chunks))]
)

print("Stored", len(chunks), "chunks in ChromaDB")
```

Why in RAG:

PDF → Chunks → Embeddings → ChromaDB → Retrieval

Embeddings are just vectors in memory right now.
ChromaDB stores them so we can later search for the most relevant chunks using a query.

Explanation:

- chromadb is a vector database.
- Client() creates a local ChromaDB instance.
- create_collection() creates a container for vectors and documents.
- embeddings.tolist() converts NumPy arrays into Python lists.
- documents=chunks stores the original text of each chunk.
- ids creates a unique identifier for every chunk.
- collection.add() saves everything into ChromaDB.

Output:

Stored 81 chunks in ChromaDB

Meaning:

Each chunk now contains:
1. Original text
2. 384-dimensional embedding
3. Unique ID

and is stored inside ChromaDB, ready for retrieval.

**Retrieval**

```python
query = "What is Logistic Regression?"

query_embedding = embedding_model.encode(query).tolist()

results = collection.query(
    query_embeddings=[query_embedding],
    n_results=3
)

print(results["documents"][0])
```

Why in RAG:

Find the most relevant chunks for a user's question.

Explanation:
- Convert the query into an embedding.
- Search ChromaDB using that embedding.
- Return the top 3 similar chunks.

Output:
3 chunks most relevant to "What is Logistic Regression?"

**Context Creation.**

```python
context = "\n\n".join(results["documents"][0])

print(context)
```

Why in RAG:
LLMs need a single text context, not a list of chunks.

Explanation:
- Takes the retrieved chunks.
- Merges them into one text block.
- Stores it in the variable context.

Output:
A single text containing all retrieved chunks.

**Mini RAG**

```python
question = "What is Logistic Regression?"

prompt = f"""
Answer the question using only the context below.

Context:
{context}

Question:
{question}
"""

print(prompt)
```

Why in RAG:
Combine retrieved context and user question into a prompt for the LLM.

Explanation:
- question stores the user's query.
- context contains the retrieved chunks.
- f-string inserts both into one prompt.
- This prompt will be sent to an LLM next.

Output:
A complete prompt containing context + question.

**send the prompt to an LLM.**

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-base")
model = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-base")

inputs = tokenizer(prompt, return_tensors="pt")

outputs = model.generate(
    **inputs,
    max_new_tokens=100
)

answer = tokenizer.decode(outputs[0], skip_special_tokens=True)

print(answer)
```

Why in RAG:

Prompt → LLM → Answer

Explanation:

- Loads FLAN-T5 tokenizer.
- Loads FLAN-T5 model.
- Converts prompt into tokens.
- Generates an answer.
- Converts output tokens back to text.

Output:

A direct answer generated from the retrieved context.
