First Chunking Example

```python
text = """
Python is a programming language.
Django is a web framework.
FastAPI is used for APIs.
Machine learning uses data.
"""

chunks = text.split(".")
for item in chunks:
    if item.strip():  # This skips the final empty item
        print(item.strip())

print(chunks)
```

```python
text = """
Python is a programming language.
Django is a web framework.
FastAPI is used for APIs.
Machine learning uses data.
"""

chunks = text.split(".")

for i, chunk in enumerate(chunks):
    print(f"Chunk {i}: {chunk}")
```

```python
chunks = [chunk.strip() for chunk in text.split(".") if chunk.strip()]
print(f"{chunks}")
```

```python
from pypdf import PdfReader

reader = PdfReader("NBayesLogReg.pdf")

print("Total Pages:", len(reader.pages))

text = ""

for page in reader.pages:
    text += page.extract_text()

print("\nTotal Characters:", len(text))
```

Explanation

from pypdf import PdfReader
- Imports the PDF reader library.

reader = PdfReader("NBayesLogReg.pdf")
- Loads the PDF file into memory.

print("Total Pages:", len(reader.pages))
- Counts the total number of pages in the PDF.

text = ""
- Creates an empty string to store extracted text.

for page in reader.pages:
- Loops through every page in the PDF.

text += page.extract_text()
- Extracts text from the current page and appends it to the text variable.

print("\nTotal Characters:", len(text))
- Displays the total number of characters extracted from the PDF.

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

print("Import successful")
```

```python
splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=100
)

chunks = splitter.split_text(text)

print("Total Chunks:", len(chunks))
```

Explanation

splitter = RecursiveCharacterTextSplitter(...)
- Creates a text splitter object.

chunk_size=500
- Maximum size of each chunk is 500 characters.

chunk_overlap=100
- Repeats 100 characters from the previous chunk.

chunks = splitter.split_text(text)
- Splits the PDF text into chunks.

print("Total Chunks:", len(chunks))
- Displays the total number of chunks created.

```python
for i, chunk in enumerate(chunks[:3]):
    print(f"\n----- Chunk {i+1} -----\n")
    print(chunk)
```

Explanation

chunks[:3]
- Selects the first 3 chunks.

enumerate(...)
- Provides chunk number and content.
